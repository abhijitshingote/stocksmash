```python
import csv,pickle
```

```python
stocknames=[]
with open('stocklist.csv', 'rt') as csvfile:
    spamreader = csv.reader(csvfile)
    for row in spamreader:
        stocknames.append(','.join([row[0][5:],row[1].split('(')[0]]))
```

```python
stocknames
```

```python
pickle.dump(stocknames,open('stocknames.pickle','wb'))
```
